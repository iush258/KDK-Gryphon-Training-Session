### Errors and Exception Handling

types of Errors

1. Semantic Errors
2. Syntax Error
3. Exception (runtime Error)

semantic error :- Error within the Logic
Syntax:- Within the syntax of error
Exception Error :- Error Occured during the Runtime of Program 


Exception Error Can be Avoided using Try and Except Block
```
try:

    <code>

except:

    <code after the exception is occured>
```


Types of Exceptions
1. ZeroDivisionError
2. IndexError
3. ValueError
4. TypeError
5. Keyerror
6. FileNotFoundError

We can Mention multiple Exception in Exception block using ,
Ex:
```
    except(ZeroDivisionError , TypeError , IndexError):
        print("Error is Handled By Try and Exception")
```





````python
try:
    a = int(input("enter 1st No:"))
    b = int(input("enter 1st No:"))
    c= a/b

except(ZeroDivisionError):
    print("Cannot divide anything using 0")


except(ValueError):
    print("Please Provide Integers only")

#else block gets executed if there is no Exception

else:
    print(c)

# finally blocks gets executed Anyways even when the Exception is occured or not

finally:
    print("Operations are Done Completed")

# Raise is used to throw and Custom Exception to the user


````
