### Ordinal Encoding for `Flat_type`

```python
# Get unique values in 'Flat_type' to define order
print("Unique Flat Types:", df4['Flat_type'].unique())

# Define the desired order for 'Flat_type'
flat_type_order = ['1 RK', '1 BHK', '2 BHK', '3 BHK']

# Initialize OrdinalEncoder with the specified order
flat_type_encoder = OrdinalEncoder(categories=[flat_type_order])

# Fit and transform the 'Flat_type' column
df4['Flat_type_encoded'] = flat_type_encoder.fit_transform(df4[['Flat_type']])

# Display the updated DataFrame with the new encoded column
display(df4.head())
```

```python
# Optionally, drop the original 'Flat_type' column if no longer needed
df4 = df4.drop('Flat_type', axis=1)
display(df4.head())
```
