```python
l=[10,20,30,"KDK","Nagpur",]
key=eval(input("Enter the Value to find:"))
f=False
for i in l:

    if (i==key):
        f=True
        break
    else:
        f=False

if f :
    print("found it")

else :
    print("not found")
```

```python
import numpy as np
# print(np.__version__)
#to check version of numpy installed

# to create 1d array
n1 = np.array([1, 2, 3, 4, 5])

# to create 2d array
n2 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

# printing dimention of array
print(n1.ndim)
print(n2.ndim)
print()


# printing shape of array
print(n1.shape)
print(n2.shape)
print()


#  to create a array from given ranges(start,stop,step) we can use
# arrange()
n3=np.arange(25)
print(n3)
print()


# ones()
# it gives arrays to 1'set
n4=np.ones((3,4),dtype=int)
print(n4)
print()
# similarly zeros()
# it gives arrays of zero

n5=np.zeros((3,4),dtype=int)
print(n5)
```
