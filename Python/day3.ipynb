* ## Modules and Packages
+ ### Modules
    Modules are the Collection of Related Functions
    
    modules is a python (.py) file and it stores interrelated functions , classes ,constants etc

    + #### Types of modules
        1. In-built modules
        2. User defined modules

    + #### In-Built Modules
        This are the preinstalled modules Installed during the installation of Python interpreter
        ex: math ,random , os ,sys , utils etc

    + #### User defined Modules
        This are the Modules Created by Users to Use them in Thier Program 


    + #### Different Ways of Importing a Module
        1. ##### Normal Import

            Syntax :
                Import < module name >
        
            ex:    
            ``` import math ```


        2. ##### Import using "from" keyword

            Instead of Importing Entire Module in the program the user can import some specific sub modules / Functions from the Modules to Avoid Using Unnesseary Memory 
    
            Syntax :
                from < module name > import < sub module / function >

            ex:
                ``` from math import log ```

        3. ##### Import Wih "as" Keyword

            "as" Keyword allows user to import modules in program by mentioning thier name with Aliases to reduce the name of the module during function calls

            Syntax :
                import < module name > as < alias >

            ex:
                ``` import matplotlib.pyplot as plt ```
