```python
#Types of Arguments

#Positional Arguments
def posargu(a,b):
    print(f"a={a}")
    print(f"b={b}")

#default Arguments
def defargu (a , b=10):
    print(f"a={a}")
    print(f"b={b}")

#keyword Argument
def keyargu(a,b,c=20):
    print(f"a={a}")
    print(f"b={b}")
    print(f"c={c}")
    

print("Positional Argumments")
posargu(50,60)
print("\n")

print("Default Arguments")
defargu(10,20)
defargu(30)
print("\n")

print("Keyword Argumments")
keyargu(90,70)

```
